```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

# Study Area

```python
# Generating map of study areas

fig, ax = plt.subplots(subplot_kw={'projection': ccrs.PlateCarree()}, figsize=(10, 15))

# Set map boundaries
ax.set_extent([-5, 2, 49, 59])

# Add natural features using cartopy
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, edgecolor='black')
ax.add_feature(cfeature.LAKES, edgecolor='black')
ax.add_feature(cfeature.RIVERS)

# Highlight study area around London and Brighton using circles
radius = 0.5  # This is in degrees; adjust as needed
circle_london = mpatches.Circle(xy=(0.118092, 51.509865), radius=radius, color='red', alpha=0.2)
circle_brighton = mpatches.Circle(xy=(-0.137163, 50.822530), radius=radius, color='blue', alpha=0.2)
ax.add_patch(circle_london)
ax.add_patch(circle_brighton)

# Add labels for clarity
ax.text(0.118092, 51.509865, 'London', horizontalalignment='center')
ax.text(-0.137163, 50.822530, 'Brighton', horizontalalignment='center')

plt.savefig("./Data/UK_study_area.png", dpi=500)
plt.show()
```
